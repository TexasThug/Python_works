## Problème du prix

Notre équipe marketing a besoin du prix moyen des livres.
- **Question** : La colonne price est de type **object**. Comment allez-vous la transformer en une colonne numérique **float** ou **int** sur laquelle on peut faire des calculs ?

```python
df_copy["price"] = df_copy['price'].str.replace(",","").str.strip()
```

```python

df_copy["price"] = df_copy['price'].str.replace("Free","0")
```

```python
df_copy["price"] = df_copy['price'].astype(float)
```

```python
df_copy["nb_ratings"] = df_copy["nb_ratings"].str.replace("stars","")
```

```python
df_copy["note"] =df_copy["stars"].str.extract(r'(\d+(?:\.\d+)?)out of 5').astype(float)
df_copy["nb_ratings"] =df_copy["stars"].str.extract(r'(\d+(?:\.\d+)?)s*ratings').astype(float)
df_copy.loc[df_copy["stars"]=="Not rated yet", ["note","nb_ratings"]] = np.nan
```

## Le Nettoyage de Texte (author & narrator)

Les colonnes author et narrator ont des préfixes inutiles.
- **Question** : Comment nettoyez-vous ces deux colonnes pour ne garder que les noms ?

```python
df_copy['author'] = df_copy['author'].str.replace("Writtenby:","")
df_copy['narrator'] = df_copy['narrator'].str.replace("Narratedby:","")
```
