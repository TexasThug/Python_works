```python
import numpy as np 
import matplotlib.pyplot as plt 
```

```python
acc = 35
time = np.arange(0,10,0.1)
vitesse = acc * time 
distance = 0.5 * acc * time**2
```

```python
plt.figure(figsize = (7,5), dpi = 70)
plt.plot(time, vitesse, c="green", label = "Vitesse")
plt.plot(time, distance, c="blue", label = "Distance")
plt.xlabel("Temps (s)")
plt.ylabel("Vitesse en m/s et Distance en m")
plt.legend()
plt.grid(visible = True)
```

```python
fig, ax1 = plt.subplots()
ax1.set_ylabel("Distance (m)")
ax1.set_xlabel("Temps (s)")
ax1.plot(time, distance, c="blue")

ax2 = ax1.twinx()

ax2.set_ylabel("Vitesse (m/s)")
ax2.set_xlabel("Temps (s)")
ax2.plot(time, vitesse, c="green")
```

```python
fig, ax1 = plt.subplots()
ax1.set_ylabel("Distance (m)")
ax1.set_xlabel("Temps (s)")
ax1.plot(time, distance, c="blue")

ax1.set_yticks(np.linspace(*ax1.get_ybound(),10))
ax1.tick_params(axis="y", labelcolor="blue")
ax1.xaxis.grid()
ax1.yaxis.grid()

ax2 = ax1.twinx()

ax2.set_ylabel("Vitesse (km/h)")
ax2.set_xlabel("Temps (s)")
ax2.plot(time, vitesse, c="green", label = "Vitesse")

ax2.set_yticks(np.linspace(*ax2.get_ybound(),10))
ax2.tick_params(axis="y", labelcolor="green")

fig.set_size_inches(7,5)
fig.set_dpi(100)
fig.legend()
```
