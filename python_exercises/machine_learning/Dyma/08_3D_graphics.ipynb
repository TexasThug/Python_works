```python
import numpy as np 
import matplotlib.pyplot as plt 
from sklearn.datasets import load_breast_cancer
```

```python
breast_cancer = load_breast_cancer()
x= breast_cancer.data
y= breast_cancer.target
names = list(breast_cancer.target_names)
```

```python
print(names)
```

```python
x.shape
```

```python
y.shape
```

```python
tumor_radius = x[:,1]
tumor_texture = x[:,2]
```

```python
np.unique(y).size
```

```python
print(y)
```

```python
plt.scatter(tumor_radius, tumor_texture, c=y, alpha = 0.8)
```

```python
ax=plt.axes(projection = "3d")
ax.scatter(tumor_radius, tumor_texture, y, c=y, alpha = 0.6, s=50)
ax.set_xlabel("Rayon de la tumeur")
ax.set_ylabel("Texture de la tumeur")
ax.legend(names)
```
