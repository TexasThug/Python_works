## Indexing


```python
A = np.array([1,2,3,4])
for i in range(len(A)):
    print(A[i])
```

```python
B = np.array([[1,2,3],[4,5,6]])
print(B)
```

## Slicing 

```python
D = np.random.randint(-100,100,size=(4,4))
print(D)
```

```python
E = D[:,-1]
print(E)
```

```python
F = D[1:3,:]
print(F)
```

```python
G = D[1:3,1:3]
print(G)
```

```python
H = D[:,0:3:2]
print(H)
```
