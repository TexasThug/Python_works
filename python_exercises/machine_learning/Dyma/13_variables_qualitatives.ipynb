```python
import matplotlib.pyplot as plt 
import seaborn as sns 
sns.set_style("darkgrid")
```

```python
penguins = sns.load_dataset("penguins")
penguins.dropna(inplace = True)
```

```python
penguins.head()
```

```python
sns.catplot(data = penguins, x = "bill_length_mm", y = "species", hue = "species")
```

```python
sns.catplot(data = penguins, x = "bill_length_mm", y = "species", kind = "box", hue = "species")
plt.show()
```

```python
sns.catplot(data = penguins, x = "bill_length_mm", y = "species", kind = "violin", hue = "sex", split = True)
```

```python
sns.catplot(data = penguins, x = "species", y = "bill_length_mm", col = "sex", kind = "bar")
plt.show()
```
