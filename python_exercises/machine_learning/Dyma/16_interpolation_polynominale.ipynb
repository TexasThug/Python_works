```python
import scipy.interpolate as interp1d
import numpy as np 
import matplotlib.pyplot as plt 
```

## 1. interpolation avec Scipy

```python
x = np.arange(2, 11, 1)
y = (np.cos(-x**2/9)+ 1)/2.9+0.7

x = np.delete(x, 6)
y = np.delete(y, 6)

plt.figure()
plt.scatter(x, y, color="blue", label = "Données")
plt.xlabel("Heure")
plt.ylabel("Taux de glucose dans le sang (g/L)")
plt.grid(visible = True)

f_1deg = interp1d(x, y)
f_2deg = interp1d(x, y, kind = "quadratic")
f_3deg = interp1d(x, y, kind = "cubic")

x2 = np.linspace(2, 10, 1000)
plt.plot(x2, f_1deg(x2), color = "green", label = "Interpolation linéaire")
plt.plot(x2, f_2deg(x2), color = "purple", label = "Interpolation quadratique", linestyle = "dashdot")
plt.plot(x2, f_3deg(x2), color = "orange", label = "Interpolation cubique", linestyle = "dashdot")
plt.legend(bbox_to_anchor = (1.5, 1.0), loc = "upper right")

print(f_1deg(8))
print(f_2deg(8))
print(f_3deg(8))
```
