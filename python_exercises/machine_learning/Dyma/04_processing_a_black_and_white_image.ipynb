## Affichage de l'image 


```python
plt.gray()
plt.imshow(ascent)
plt.show()

ascent.shape
print(ascent)
```

```python
zoom_ascent = np.copy(ascent[200:400,350:])
print(zoom_ascent)
plt.imshow(zoom_ascent)

zoom_ascent[zoom_ascent>150] = 255
plt.imshow(zoom_ascent)

zoom_ascent[zoom_ascent<100] = 0
plt.imshow(zoom_ascent)

plt.imshow(ascent)
M = np.full((ascent.shape[0],ascent.shape[1]),255)
print(M)

inverted_ascent = M-ascent
plt.imshow(inverted_ascent)
```
