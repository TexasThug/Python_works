```python
import matplotlib.pyplot as plt 
import seaborn as sns 
sns.set_style("darkgrid")
```

```python
iris = sns.load_dataset("iris")
iris.head()
```

```python
sns.histplot(data= iris, x = "sepal_width", stat = "density", kde = True)
```

```python
sns.kdeplot(data = iris, x = "sepal_width", fill = True, hue = "species" )
```

```python
sns.kdeplot(data = iris, x = "sepal_width", y = "petal_width", fill = True, cbar = True )
```

```python
#2. Pairplot 
```

```python
sns.pairplot(iris, dropna = True, hue = "species")
```
