```python
import numpy as np
```

## 1. Addition et soustraction

```python
A = np.array([[-2,-3,0], [1,2,3]])
print(A)
```

```python
n_lin = A.shape[0]
n_col = A.shape[1]
```

```python
B = np.ones((n_lin, n_col))
print(B)
```

```python
C = A + B
print(C)
```
