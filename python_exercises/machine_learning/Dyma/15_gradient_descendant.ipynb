```python
import numpy as np 
import matplotlib.pyplot as plt 
import scipy as sp
plt.style.use("ggplot")
```

## 1. Création de la fonction du 2ème degré 

```python
f = lambda x : x**2 + 5 * x - 10
f_deriv = lambda x: 2*x +5

x = np.arange(-20,15,0.1)
plt.plot(x, f(x))
```

## 2. Algorithme du gradient descendant

```python
def grad_desc(x, f_deriv, lrn_rate, epsilon) : 
    x_arch = []
    n_iter = 0
    while (abs(f_deriv(x)) > epsilon): 
        x_arch.append(x)
        n_iter =+ 1 
        x = x - lrn_rate*f_deriv(x)
    return x, n_iter, x_arch
```

```python
x_min, n_iter, x_arch = grad_desc(-20, f_deriv, 0.01, 0.00001)
y_min = f(x_min)
print(f"Le minimum se trouve en ({x_min}, {y_min})")
```

```python
plt.figure()
plt.plot(x, f(x))
plt.scatter(x_min, y_min, color ="blue")
```

```python
plt.figure()
for i in range(n_iter): 
    plt.scatter(x_arch[i], f(x_arch[i]))
```

## 3. Minimisation avec Scipy 

```python
x_min = sp.optimize.minimize(f, -20, options ={"gtol" : 0.001, "disp" : True}) 
```
